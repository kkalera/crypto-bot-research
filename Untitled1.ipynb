```python
%load_ext autoreload
%autoreload 2

import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.optimizers import RMSprop, Adam, SGD
from keras import backend
import pandas as pd
import matplotlib.pyplot as plt
import trading
from DQNAgent import DQNAgent
import time
import random
from numba import jit
from collections import deque
```

```python
def create_model(input_shape, output_num):
    backend.clear_session()
    model = Sequential()
    
    model.add(Dense(5, input_shape=input_shape, activation="relu"))
    #model.add(Dropout(0.3))
    model.add(Dense(5, activation="relu"))
    #model.add(Dropout(0.3))
    #model.add(Dense(30, activation="relu"))
    #model.add(Dropout(0.3))
    model.add(Dense(output_num, activation="linear"))
    
    #opt = Adam(decay = 1e-6)
    #opt = SGD(lr=0.01, decay = 1e-6, momentum=0.9, nesterov=True)
    model.compile(optimizer="adam", loss="mse")
    
    model.summary()
    return model
```

```python
data = np.sin(np.arange(200)/30.0)
#data = np.arange(200/5.0) #linearly increasing prices
plt.plot(data)
```

```python
#env = trading.SimpleTestEnvironment()
env = trading.NewTradingEnv(data=data)

length_of_day = 3
#env.day_len = length_of_day
env.state_len = length_of_day

input_data_shape = env.reset()[0].shape
print(input_data_shape)
action_size = 3

model = create_model(input_data_shape, action_size)
agent = DQNAgent(model, action_size)
```

```python
@jit
def replay_random(memory_x, memory_y, model, replay_size=250):
    
    zipped_list = list(zip(memory_x, memory_y))
    mem_len = len(list(memory_x))
    sample_len = replay_size if (mem_len > replay_size) else mem_len
    samples = random.sample(zipped_list, sample_len)
    
    for i in range(sample_len):
        model.fit(samples[i][0], samples[i][1], batch_size=1, epochs=1, verbose=0)
    
    #model.fit(samples_x, samples_y)
```

```python
@jit
def train(epochs = 100, reset_model = False):
    if reset_model:
        model = create_model(input_data_shape, action_size)
    
    env.data = data
    epsilon = 0.3
    min_epsilon = 0.05
    epsilon_decay = 0.00005
    gamma = 0.9
    count = 0
    
    x_data = []
    y_data = []
    
    wallets = []


    for ep in range(epochs):
        count +=1
        state, score, done = env.reset()    

        while not done:
            q_val = model.predict(state.reshape(1, length_of_day+1), batch_size = 1)

            # Choose an action
            if (random.random() < epsilon):
                action = np.random.randint(0,3)
                decayed = epsilon - epsilon_decay
                epsilon = decayed if decayed > min_epsilon else min_epsilon
            else:
                action = np.argmax(q_val)

            # Take the action and get the new state, score, and done values
            next_state, next_score, done = env.step(action)

            new_qval = model.predict(next_state.reshape(1, length_of_day+1), batch_size=1)
            max_qval = np.amax(new_qval)

            y = np.zeros((1, 3))
            y[:] = q_val[:]

            if not done:
                update = next_score + (gamma * max_qval)
            else:
                update = next_score

            y[0][action] = next_score
            model.fit(state.reshape(1, length_of_day+1), y, batch_size=1, epochs=1, verbose=0)
            
            x_data.append(state.reshape(1, length_of_day+1))
            y_data.append(y)

            state = next_state 
            
        replay_random(x_data, y_data, model)
        #wallets.append(env.wallet)
        wallets.append(env.balance)
        #print("Epoch: {}, final wallet: {}, epsilon: {} ".format(ep+1, env.wallet, epsilon))
        print("Epoch: {}, final wallet: {}, epsilon: {} ".format(ep+1, env.balance, epsilon))
    

    env.render()
    plt.plot(wallets)
    plt.show()
    model.save("test_trader")
```
